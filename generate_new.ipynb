```python
from unsloth import FastLanguageModel
from transformers import AutoTokenizer
from peft import PeftModel

# Load the BASE model (Unsloth's Llama-3 3B)
model = FastLanguageModel.from_pretrained(
    model_name="unsloth/Llama-3.2-3B",
    max_seq_length=4096,  # Use the same sequence length as training
    dtype=None  # Automatically selects bf16 or fp16 based on GPU
)

# Load the LoRA Adapters from Checkpoint
model = PeftModel.from_pretrained(model, "outputs/checkpoint-2500")

# Load Tokenizer
tokenizer = AutoTokenizer.from_pretrained("unsloth/Llama-3.2-3B")
```
