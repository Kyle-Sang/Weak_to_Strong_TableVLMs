```python
import json
import pandas as pd

with open('./tabFact_data/tabFact.json', 'r') as f:
    data = json.load(f)

formatted_data = []
data = data['train']  # Access the training split
data = data[len(data)//4:]

num_samples = len(data) // 4
### Randomly sample 1/4 of the data
synthetic_quarter = np.random.choice(data, num_samples, replace=False)

## Put all the examples not in synthetic_quarter in a separate variable
remaining_data = [item for item in data if item not in synthetic_quarter]

### Access two random examples from the remaining_data pool
x, y = np.random.choice(remaining_data, 2, replace=False)
```

```python
from tqdm import tqdm
synthetic_data = []
for sample in tqdm(synthetic_quarter):
    examples = np.random.choice(remaining_data, 2, replace=False)
    
    table = sample['tables']['bracket']
    question = sample['question']
    input = f"{question}\n{table}"
    text = ""

    for example in examples:
        example_table = example['tables']['bracket']
        example_question = example['question']
        input_example = f"\n{example_question}\n{example_table}"
        text += f"### Question: {input}\n ### Answer: {example['answer']}\n"

    text = f"""Please read the following table, the texts before the table, 
    and the text after the table. Answer the question with an output value 
    on the likelihood from 0 (false) to 1 (true), ### Question: {input}\n ### Answer: """

    # Encode the input text and move it to GPU
    input_ids = tokenizer.encode(text, return_tensors="pt").to("cuda")

    # Generate output using the model
    length = input_ids.size()[1] + 1
    output_ids = model.generate(input_ids, max_length=length)

    # Decode the output to text
    output_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)
    if output_text[-1].isnumeric() and (int(output_text[-1]) == 1 or int(output_text[-1] == 0)):
        sample['answer'] = output_text[-1]
    else:
        sample['answer'] = "INVALID"
    synthetic_data.append(sample)

import os
json_path = os.path.join("./tabFact_pseudolabels/", "tabfact_pseudolabels.json")
with open(json_path, "w", encoding="utf-8") as f:
    json.dump(synthetic_data, f, indent=4, ensure_ascii=False)
```
